## Definitions

From my Alma Mater (https://www.sas.upenn.edu/~haroldfs/dravling/grice.html):

**The maxim of quantity**, where one tries to be as informative as one possibly can, and gives as much information as is needed, and no more.

**The maxim of quality**, where one tries to be truthful, and does not give information that is false or that is not supported by evidence.

**The maxim of relation**, where one tries to be relevant, and says things that are pertinent to the discussion.

**The maxim of manner**, when one tries to be as clear, as brief, and as orderly as one can in what one says, and where one avoids obscurity and ambiguity.

As the maxims stand, there may be an overlap, as regards the length of what one says, between the maxims of quantity and manner; this overlap can be explained (partially if not entirely) by thinking of the maxim of quantity (artificial though this approach may be) in terms of units of information. In other words, if the listener needs, let us say, five units of information from the speaker, but gets less, or more than the expected number, then the speaker is breaking the maxim of quantity. However, if the speaker gives the five required units of information, but is either too curt or long-winded in conveying them to the listener, then the maxim of manner is broken. The dividing line however, may be rather thin or unclear, and there are times when we may say that both the maxims of quantity and quality are broken by the same factors.

## Our version of the Maxim of Manner

The standard tasks that I experiment are all multiple choice questions which allows for maximally terse few token responses, '(A)' or 'Yes'. This allows for a maximal brevity and/or clarity with length limits or restricting answers to schemas. 

1. One or a few tokens via prefix (run tokenizers)
2. Schema, either suggested or enforced

```python

import os
import tarfile
import pandas as pd
import glob
import json

run_dir = 'grice_runs'
if not os.path.exists(run_dir):
    with tarfile.open('runs.tgz', 'r:gz') as tar:
        tar.extractall()

experiments = glob.glob(f'{run_dir}/*/*.csv')
data_df = pd.DataFrame()
for exp_run in experiments:
    exp_run_df = pd.read_csv(exp_run)
    print(exp_run)
    data_df = pd.concat([data_df, exp_run_df])

data_df['task_config'] = data_df['task_config'].apply(json.loads)
data_df['model_config'] = data_df['model_config'].apply(json.loads)
display(data_df)
```

```python
import re
import importlib
import os
import sys
sys.path.append(os.path.join(os.getcwd(),"..", ".."))

def add_label(row):
   model_config = row['model_config']
   if 'prefix' not in model_config:
       return "Unconstrained"
   prefix_val = model_config['prefix']
   if 'Please answer the following question while adhering to Gricean Maxims: Particularly the Maxim of Manner: Be clear—avoid obscurity and ambiguity, be brief and orderly. ' in prefix_val:
       return "Grice"
   if 'Please answer the following question, this is a multiple choice question and there is no benefit to the grade in showing your work.' in prefix_val:
       return "No Grice"
   if 'The output should be formatted as a JSON instance that conforms to the JSON schema below' in prefix_val:
       return "Schema"

data_df['task/model'] = data_df.apply(lambda x: add_label(x), axis=1)
display_df = data_df

for model_task in display_df['task/model'].unique():
    last_run = display_df[display_df['task/model'] == model_task]['run'].max()
    determinism_counts = []
    for i in range(0, last_run + 1):
       determinism_counts.append(len(display_df[(display_df['run'] == i)
                                 & (display_df['task/model'] ==  model_task)].index))
    print(model_task, determinism_counts)


def score(row, task_module):
    try:
        parsed_answer = task_module.answer_fn(row, row['task_config'])
        if parsed_answer is None:
            return False
    except LookupError as e:
        return False
    return task_module.correct_fn(row, row['task_config'])

for task_name in display_df['task'].unique():
    task_module = importlib.import_module(f'tasks.{task_name}')
    display_df['correct'] = display_df.apply(lambda x: score(x, task_module), axis=1)

display(display_df)
```

# Accuracy impact of applying Maxim of Manner

A general rule of thumb is that encouraging the LLM to go for longer, COT (chain of thought) techniques do this, improves performance. Lets find out:

```python
import plotnine as p9
from mizani.formatters import percent_format

result_df = (
    display_df
    .groupby(['task/model', 'run'])['correct']
    .mean()
    .reset_index()
    .rename(columns={'correct': 'correct %'})
)



count_df = (
    display_df
    .groupby(['task/model', 'run'])
    .size()
    .reset_index(name='count')
)

max_count = count_df['count'].max()
count_df['scaled_count'] = count_df['count'] / max_count  # scaled 0–1
count_df['scaled_count'] *= 100  # match % scale of 'correct %'


result_df['config'] = result_df['task/model']  

plot_df = pd.merge(result_df, count_df, on=['task/model', 'run'], how='left')

sub_plot_df = plot_df[plot_df['run'] < 6]

color="#b300b3"
plot = (
    p9.ggplot(sub_plot_df, p9.aes(x='run', y='correct %', color='config'))
    + p9.geom_line()
    + p9.geom_point(p9.aes(size='count'))  # <- scale size to data volume
    + p9.scale_color_brewer(type='qual', palette='Set1')
    + p9.scale_y_continuous(labels=percent_format())
    + p9.scale_size(range=(1, 10))  # control min/max dot size
    + p9.labs(
        title='Model Accuracy Over Runs (Dot Size = Determ Resp @ run)',
        x='Run',
        y='Correct (%)',
        color='config',
        size='Count'
    )
    + p9.theme(
        axis_text_x=p9.element_text(),
        text=p9.element_text(color=color),
        axis_text=p9.element_text(color=color),
    )
)


display(plot)
```

```python
import pandas as pd
import tiktoken
from transformers import AutoTokenizer

# Function to choose tokenizer based on model name
def get_tokenizer(model_name):
    if 'gpt-35' in model_name or 'gpt-4o' in model_name:
        try:
            return tiktoken.encoding_for_model(model_name)
        except KeyError:
            return tiktoken.get_encoding("cl100k_base")
    else:
        # Fallback tokenizer for unsupported or private models
        try:
            return AutoTokenizer.from_pretrained(model_name, use_fast=True)
        except Exception as e:
            print(f"Warning: Failed to load tokenizer for {model_name}, using fallback. Error: {e}")
            return AutoTokenizer.from_pretrained("bert-base-uncased", use_fast=True)


# Cache tokenizers to avoid re-loading for repeated models
tokenizer_cache = {}

def tokenize_length(text, model_name):
    if model_name not in tokenizer_cache:
        tokenizer_cache[model_name] = get_tokenizer(model_name)
    
    tokenizer = tokenizer_cache[model_name]

    # tiktoken vs HF interface difference
    if isinstance(tokenizer, tiktoken.Encoding):
        return len(tokenizer.encode(text))
    else:
        return len(tokenizer.encode(text, add_special_tokens=False))

# Apply to dataframe
display_df['question token length'] = display_df.apply(
    lambda row: tokenize_length(row['question'], row['model']),
    axis=1
)

display_df['response token length'] = display_df.apply(
    lambda row: tokenize_length(row['response'], row['model']),
    axis=1
)

display(display_df)
```

```python
import pandas as pd

# Start with only run 0
round_0_df = display_df[display_df['run'] == 0].copy()

# Optional: bin token length (e.g., into bins of 10)
round_0_df['length_bin'] = pd.cut(
    round_0_df['response token length'],
    bins=list(range(10))
#    labels=['0–5', '6–10', '11–20', '21–50', '51–100', '100+']
)

# Group by task/model and token length bin, collect responses and their counts
def response_count_set(series):
    return series.value_counts().to_dict()

binned_df = (
    round_0_df
    .groupby(['task/model', 'length_bin'], observed=False)['response']
    .agg(response_count_set)
    .reset_index()
    .rename(columns={'response': 'response_counts'})
)

display(binned_df)
# Show the resulting dataframe
#import ace_tools as tools; tools.display_dataframe_to_user(name="Binned #Response Counts", dataframe=binned_df)
```

```python
grice_df = round_0_df[round_0_df['task/model'] == 'Grice'].copy()
no_grice_df = round_0_df[round_0_df['task/model'] == 'No Grice'].copy()
schema_df = round_0_df[round_0_df['task/model'] == 'Schema'].copy()
unconst_df = round_0_df[round_0_df['task/model'] == 'Unconstrained'].copy()


# Define exponential bins
bins = [0, 1, 10, 100, float('inf')]
labels = ['(0–1]', '(1–10]', '(10–100]', f'(100–inf]']

# Apply binning
grice_df['length_bin'] = pd.cut(grice_df['response token length'], bins=bins, labels=labels, right=True)
no_grice_df['length_bin'] = pd.cut(no_grice_df['response token length'], bins=bins, labels=labels, right=True)
schema_df['length_bin'] = pd.cut(schema_df['response token length'], bins=bins, labels=labels, right=True)
unconst_df['length_bin'] = pd.cut(unconst_df['response token length'], bins=bins, labels=labels, right=True)


# Combine and mark source
grice_df['group'] = 'Grice'
no_grice_df['group'] = 'No Grice'
schema_df['group'] = 'Schema'
unconst_df['group'] = 'Unconstrained'

#plot_df = pd.concat([grice_df, no_grice_df])
plot_df = pd.concat([grice_df, no_grice_df, schema_df, unconst_df])


import plotnine as p9

plot = (
    p9.ggplot(plot_df, p9.aes(x='length_bin', fill='group'))
    + p9.geom_bar(position='dodge', color='black')
    + p9.labs(
        title='Token Length Distribution (Exponential Bins)',
        x='Number of Tokens in Response (binned)',
        y='Count',
        fill='Prompt Type'
    )
    + p9.theme(
        axis_text_x=p9.element_text(),
        text=p9.element_text(color=color),
        axis_text=p9.element_text(color=color),
        plot_background=p9.element_rect(fill='none'),
        panel_background=p9.element_rect(fill='none'),
        panel_grid_minor_x=p9.element_blank(),
        panel_grid_minor_y=p9.element_blank(),
        panel_grid_major_x=p9.element_blank(),
        panel_grid_major_y=p9.element_blank()
    )
)

display(plot)
```
